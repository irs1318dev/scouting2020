```python
import pandas as pd
import bokeh.plotting
import bokeh.models as bmodels
import bokeh.models.widgets as bmw
import bokeh.layouts as blt
import bokeh.plotting as plt
import bokeh.palettes as bpalettes
import bokeh.transform as btransform
import bokeh.io

import viewer_app.data_source as va_data_source 
data = va_data_source.DataSource(event='test_event_2', season='2020')

bokeh.io.output_notebook()
```

```python
dft = data.teams.copy()
dft = dft.rename(columns={'name': 'team'})
dft = dft.set_index(['team'])
df = dft.filter(['team', 'matches_played'])
dftotal = data.enum_measures.copy()
dfauto = dftotal[dftotal.phase == 'auto']
shootertasks = ['launchLower', 'launchOuter']
dfautotask = dfauto[dfauto.task.isin(shootertasks)]
dfautotask = dfautotask.groupby(['team', 'task' ]).sum()
dfautotask = dfautotask.filter(['team', 'task', 'successes'])
dfautotask = dfautotask.unstack()
dfautotask.columns = dfautotask.columns.droplevel()
grouped = dftotal.groupby(['team', 'task' ]).sum()
total = grouped.filter(['team', 'task', 'successes'])
total = total.unstack()
total.columns = total.columns.droplevel()
total['matches'] = df['matches_played']
total['autoLower'] = dfautotask['launchLower']
total['autoOuter'] = dfautotask['launchOuter']
total = total.fillna(0)
total['teleLower'] = total['launchLower'] - total['autoLower']
total['teleOuter'] = total['launchOuter'] - total['autoOuter']
total['movePoints'] = total['movedAuto'] * 5
total['autoLowerPoints'] = total['autoLower'] * 2
total['autoOuterPoints'] = total['autoOuter'] * 4
total['teleLowerPoints'] = total['teleLower'] * 1
total['teleOuterPoints'] = total['teleOuter'] * 2
total['climb'] = total['climbPosition_Center'] + total['climbPosition_Side']
total['climbpoints'] = total['climb'] * 25
total['parkpoints'] = total['climbPosition_Park'] * 5
total['positionControlPoints'] = total['positionControl'] * 25
total['rotationControlPoints'] = total['rotationControl'] * 15
total['points'] = (total['autoLowerPoints'] + total['autoOuterPoints'] + total['teleLowerPoints'] +
                   total['teleOuterPoints'] + total['climb'] + total['climbpoints'] + total['parkpoints'] +
                   total['positionControlPoints'] + total['rotationControlPoints'])
average = total.div(total.matches, axis=0)
for x in range(len(average.columns)):
    average.rename(columns={ average.columns[x]: ('avg_' + average.columns[x])}, inplace = True)
complete = total.merge(average, left_index=True, right_index=True)
complete = complete.sort_values(by ='avg_points' )
complete
```

```python
Rank_cds = bmodels.ColumnDataSource(complete)
fixed2 = bmw.NumberFormatter(format='0.00')
cols = [
        bmw.TableColumn(field='team', title='Team'),
        bmw.TableColumn(field='avg_points', title='Average Points', formatter=fixed2),
        bmw.TableColumn(field='autoLower', title='Total Shoot Lower Auto', formatter=fixed2),
        bmw.TableColumn(field='autoOuter', title='Total Shoot Upper Auto', formatter=fixed2),
        bmw.TableColumn(field='teleLower', title='Total Shoot Lower Teleop', formatter=fixed2),
        bmw.TableColumn(field='teleOuter', title='Total Shoot Upper Teleop', formatter=fixed2),
        bmw.TableColumn(field='launchLower', title='Total Shoot Lower', formatter=fixed2),
        bmw.TableColumn(field='launchOuter', title='Total Shoot Upper', formatter=fixed2),
        bmw.TableColumn(field='positionControl', title='Total position control', formatter=fixed2),
        bmw.TableColumn(field='rotationControl', title='Total Rotation Control', formatter=fixed2),
        bmw.TableColumn(field='movedAuto', title='Total moved auto', formatter=fixed2),
        bmw.TableColumn(field='climbPosition_Center', title='Total Climbs', formatter=fixed2)
     ]
data_table = bmw.DataTable(source=Rank_cds, columns=cols, width=1400, height=380)
bokeh.io.show(data_table)
```

```python
df_temp = complete
df_new = df_temp.filter(['avg_autoLowerPoints', 'avg_autoOuterPoints', 'avg_teleLowerPoints',
                         'avg_teleOuterPoints', 'avg_climbpoints', 'avg_parkpoints', 
                         'avg_positionControlPoints', 'avg_rotationControlPoints'], axis=1)
points_cds = bmodels.ColumnDataSource(df_new)
tm_col_name = points_cds.column_names[0]
task = points_cds.column_names[1:]
tooltips = [
    ("", "team: @team"),
    ("", "$name: @$name")
]
p = plt.figure(title='Points Chart', x_range=points_cds.data[tm_col_name],
               plot_width=1100, plot_height=350, tooltips=tooltips, toolbar_location="above")
hr = p.vbar_stack(task, x=tm_col_name, width=0.5,
                  source=points_cds, color=bpalettes.RdBu8)
legend = bokeh.models.Legend(items=[(x, [hr[task.index(x)]]) for x in task], location=(0, 0))
p.add_layout(legend, 'right')
p.xaxis.major_label_orientation = 3.14 / 4
bokeh.io.show(p)
```
